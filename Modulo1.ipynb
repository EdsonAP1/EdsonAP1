```python
intentos = 0

# El bucle controla que no nos pasemos de 3 intentos
while intentos < 3:
    contraseña = input("Introduce la contraseña: ")
    intentos = intentos + 1  # Sumamos un intento cada vez que escribe

    if contraseña == "hackme":
        print("ACCESO AUTORIZADO")
        break  # Rompe el bucle 'while' de inmediato si acertó
    else:
        print("Contraseña incorrecta")

# Fuera del bucle (sin espacios al inicio) revisamos si perdió por agotar intentos
if intentos == 3 and contraseña != "hackme":
    print("¡SISTEMA BLOQUEADO por seguridad!")
```

```python
def convertir_a_seguro(url):
  if "http://" in url: #La forma fácil: Usar la palabra clave in (que busca si un texto está metido dentro de otro)
    return("URL Insegura, se recomienda cambiar a HTTPS")
  else:
   return("URL es seguro tranquilo")
#Termina la funcion
direccion=input("Ingrese el URL:")
coreccion=convertir_a_seguro(direccion)
print(f"DIRECCION {coreccion}")

```

```python
def convertir_a_seguro(url):
    # .startswith() devuelve True si el texto empieza con "http://"
    if url.startswith("http://"):
        return "URL Insegura, se recomienda cambiar a HTTPS"
    else:
        return "URL es seguro tranquilo"

direccion = input("Ingrese el URL: ")
coreccion = convertir_a_seguro(direccion)
print(f"DIRECCION {coreccion}")
```
